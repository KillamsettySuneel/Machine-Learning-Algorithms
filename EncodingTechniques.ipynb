# Label Encoder

```python
import pandas as pd
```

```python
data = pd.read_csv(r"Color.csv")
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
Le = LabelEncoder()
data['Color'] = Le.fit_transform(data['Color'])
data.head(15)
```

# One Hot Encoder

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv(r"Color.csv")
data.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(handle_unknown = 'ignore')
encoder_data = pd.DataFrame(encoder.fit_transform(data[['Color']]).toarray())
#final_data = pd.concat([data,encoder_data],axis = 1)
final_data = data.join(encoder_data)
final_data.columns = ['Color','is_Blue','is_Green','is_Red','is_Yellow']
final_data.head(15)
```

# One Hot Encoder - Another Way

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv(r"Color.csv")
data.head()
```

```python
#data = pd.DataFrame(data)
data_one_hot_encoded = pd.get_dummies(data = data, columns = ['Color'], prefix = 'is')
#final_data = pd.concat([data, data_one_hot_encoded],axis = 1)
final_data = data.join(data_one_hot_encoded)
final_data.head(15)
```
