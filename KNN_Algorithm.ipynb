```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv(r"Churn_Modelling.csv")
data.head()
```

```python
data.info()
```

```python
data.shape
```

```python
data = data.drop(['RowNumber','CustomerId','Surname'],axis = 1)
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
Le = LabelEncoder()
data['Geography'] = Le.fit_transform(data['Geography'])
data['Gender'] = Le.fit_transform(data['Gender'])
data.head()
```

```python
data.describe()
```

```python
#Splitting dataset into training and testing dataset

x = data.drop(['Exited'],axis = 1)
y = data['Exited']

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)
```

```python
#Performing Feature Scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Build the model

from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors = 5)
model = model.fit(X_train,Y_train)
Y_pred = model.predict(X_test)
```

```python
#Performance Evaluation

from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
print("Confusion Matrix:")
print(confusion_matrix(Y_test,Y_pred))
print("Classification Report:")
print(classification_report(Y_test,Y_pred))
print("Accuracy Score:",round(accuracy_score(Y_test,Y_pred)*100,2),'%')
```

```python
#Predicting result for single observation
#Correct Prediction

new_df = pd.DataFrame({'CreditScore':502,'Geography':0,'Gender':0,'Age':42,'Tenure':8,'Balance':159660.80,'NumOfProducts':3,'HasCrCard':1,'IsActiveMember':0,'EstimatedSalary':113931.57},index = [0])
Y_pred1 = model.predict(sc.transform(new_df))
print(Y_pred1)
```
